```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

cambien1 = ctrl.Antecedent(np.arange(-1,201,0.1),'cambien1')
cambien2 = ctrl.Antecedent(np.arange(-1,201,0.1),'cambien2')
cambien3 = ctrl.Antecedent(np.arange(-1,201,0.1),'cambien3')
#dongco1 = ctrl.Antecedent(np.arange(-101,101,1),'dongco1')
#dongco2 = ctrl.Antecedent(np.arange(-101,101,1),'dongco2')
huongdi = ctrl.Consequent(np.arange(-181,181,1),'huongdi')

#dongco1['am'] = fuzz.trimf(dongco.universe,[-100,-100,0])
#dongco1['khong'] = fuzz.trimf(dongco.universe,[0,0,0])
#dongco['duong'] = fuzz.trimf(dongco.universe,[0,100,100])
#dongco1.view()

#dongco2['am'] = fuzz.trimf(dongco.universe,[-100,-100,0])
#dongco2['khong'] = fuzz.trimf(dongco.universe,[0,0,0])
#dongco2['duong'] = fuzz.trimf(dongco.universe,[0,0,100])
#dongco2.view()








```

```python
cambien1['gan'] = fuzz.trimf(cambien1.universe,[0,0,50])
cambien1['tb'] = fuzz.trimf(cambien1.universe,[50,100,200])
cambien1['xa'] = fuzz.trimf(cambien1.universe,[100,200,200])
cambien1.view()
```

```python
cambien2['gan'] = fuzz.trimf(cambien2.universe,[0,0,50])
cambien2['tb'] = fuzz.trimf(cambien2.universe,[50,100,200])
cambien2['xa'] = fuzz.trimf(cambien2.universe,[100,200,200])
cambien2.view()
```

```python
cambien3['gan'] = fuzz.trimf(cambien3.universe,[0,0,50])
cambien3['tb'] = fuzz.trimf(cambien3.universe,[50,100,200])
cambien3['xa'] = fuzz.trimf(cambien3.universe,[100,200,200])
cambien3.view()
```

```python
huongdi['trai'] = fuzz.trimf(huongdi.universe,[-180,-180,0])
huongdi['dungim'] = fuzz.trimf(huongdi.universe,[0,0,0])
huongdi['giua'] = fuzz.trimf(huongdi.universe,[-180,0,180])
huongdi['phai'] = fuzz.trimf(huongdi.universe,[0,180,180])
huongdi.view()
```

```python
rule1 = ctrl.Rule(cambien1['gan']&cambien2['gan']&cambien3['gan'],huongdi['dungim'])
rule2 = ctrl.Rule(cambien1['gan']&cambien2['gan']&cambien3['tb'],huongdi['giua'])
rule3 = ctrl.Rule(cambien1['gan']&cambien2['gan']&cambien3['xa'],huongdi['giua'])
rule4 = ctrl.Rule(cambien1['gan']&cambien2['tb']&cambien3['gan'],huongdi['phai'])
rule5 = ctrl.Rule(cambien1['gan']&cambien2['xa']&cambien3['gan'],huongdi['phai'])
rule6 = ctrl.Rule(cambien1['tb']&cambien2['gan']&cambien3['gan'],huongdi['trai'])
rule7 = ctrl.Rule(cambien1['xa']&cambien2['gan']&cambien3['gan'],huongdi['trai'])
rule8 = ctrl.Rule(cambien1['tb']&cambien2['tb']&cambien3['tb'],huongdi['giua'])
rule9 = ctrl.Rule(cambien1['tb']&cambien2['tb']&cambien3['gan'],huongdi['trai'])
rule10 = ctrl.Rule(cambien1['tb']&cambien2['tb']&cambien3['xa'],huongdi['giua'])
rule11 = ctrl.Rule(cambien1['tb']&cambien2['gan']&cambien3['tb'],huongdi['trai'])
rule12 = ctrl.Rule(cambien1['tb']&cambien2['xa']&cambien3['tb'],huongdi['phai'])
rule13 = ctrl.Rule(cambien1['gan']&cambien2['tb']&cambien3['tb'],huongdi['giua'])
rule14 = ctrl.Rule(cambien1['xa']&cambien2['tb']&cambien3['tb'],huongdi['giua'])
rule15 = ctrl.Rule(cambien1['xa']&cambien2['xa']&cambien3['xa'],huongdi['giua'])
rule16 = ctrl.Rule(cambien1['xa']&cambien2['xa']&cambien3['gan'],huongdi['giua'])
rule17 = ctrl.Rule(cambien1['xa']&cambien2['xa']&cambien3['tb'],huongdi['giua'])
rule18 = ctrl.Rule(cambien1['xa']&cambien2['gan']&cambien3['xa'],huongdi['giua'])
rule19 = ctrl.Rule(cambien1['xa']&cambien2['tb']&cambien3['xa'],huongdi['giua'])
rule20 = ctrl.Rule(cambien1['gan']&cambien2['xa']&cambien3['xa'],huongdi['giua'])
rule21 = ctrl.Rule(cambien1['tb']&cambien2['xa']&cambien3['xa'],huongdi['giua'])
rule22 = ctrl.Rule(cambien1['gan']&cambien2['xa']&cambien3['tb'],huongdi['giua'])
rule23 = ctrl.Rule(cambien1['gan']&cambien2['tb']&cambien3['xa'],huongdi['giua'])
rule24 = ctrl.Rule(cambien1['tb']&cambien2['gan']&cambien3['xa'],huongdi['giua'])
rule25 = ctrl.Rule(cambien1['xa']&cambien2['gan']&cambien3['tb'],huongdi['giua'])
rule26 = ctrl.Rule(cambien1['tb']&cambien2['xa']&cambien3['gan'],huongdi['phai'])
rule27 = ctrl.Rule(cambien1['xa']&cambien2['tb']&cambien3['gan'],huongdi['phai'])
```

```python
huongdiOn_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25,rule26,rule27])
huongdiOn=ctrl.ControlSystemSimulation(huongdiOn_ctrl)
huongdiOn.input['cambien1']=50
huongdiOn.input['cambien2']=50
huongdiOn.input['cambien3']=50
huongdiOn.compute()
print('huongdi value:',huongdiOn.output['huongdi'])
huongdi.view(sim=huongdiOn)
```
